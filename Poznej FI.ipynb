```python
#Constants
pi = 3.14
rgbcolors  = [(0,1,0),(1,1,0),(1,0,1),(0,1,1)]
import itertools


class MyPoint:
        def __init__(self, value, n):
            self.coord = (cos(2*pi*value/n),sin(2*pi*value/n))   
            self.poincoord = (1.1*cos(2*pi*value/n),1.1*sin(2*pi*value/n)) 
            self.textcoord = (1.2*cos(2*pi*value/n),1.2*sin(2*pi*value/n)) 
            self.value = value    
            self.text = ''
            self.arrows = []
        
        def gobject(self, color = (1,0,0)):
            return point(self.coord, rgbcolor = color ) + text(self.value, self.poincoord, rgbcolor = (1,0,0)) + text(self.text, self.textcoord, rgbcolor = (0,0,1),  horizontal_alignment = 'left')
    
class Zngraphic:    
    def __init__(self, n):
        self.n = n
        self.numbers = []
        self.mod = Integers(n)
        self.Znpoints = [MyPoint(i,n) for i in range(n)]
        self.color_gen = itertools.cycle(rgbcolors)
        self.color = next(self.color_gen)
        self.values_lists = []
        
    def visual(self):
        self.G = Graphics()
        for point in self.Znpoints:
            self.G = self.G + point.gobject()
        for i in range(len(self.values_lists)):
            self.color = next(self.color_gen)
            for j in range(len(self.values_lists[i]) - 1):
                self.G += self.gen_arrow(self.values_lists[i][j], self.values_lists[i][j+1])
        self.G.show(dpi=100, axes=False,aspect_ratio=1)
   
    def add_value(self, value, text = ''):
        self.Znpoints[value].text += '  ' + text
    
    def add_arrows(self, values):
        self.values_lists.append(values)
            
    def gen_arrow(self, value1, value2):
        #print(self.Znpoints[value1].coord, self.Znpoints[value2].coord, rgbcolor )
        return arrow(self.Znpoints[value1].coord, self.Znpoints[value2].coord, color = self.color) 
            
            
n = 11
g = Zngraphic(n)
g.add_value(10,'20')
g.add_arrows([2,4,8,3,6,10])
g.visual()


```
